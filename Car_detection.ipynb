```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(image_gray)
```

```python
car_detector = cv2.CascadeClassifier('/content/drive/MyDrive/Comp Vision/Cascades-20210601T134326Z-001.zip (Unzipped Files)/Cascades/cars.xml')
```

```python
car_detection = car_detector.detectMultiScale(image_gray)
```

```python
car_detection
```

```python
for (x, y, w, h) in car_detection:
  cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
cv2_imshow(image)
```

```python
image = cv2.imread('/content/drive/MyDrive/Comp Vision/Images-20210601T134554Z-001.zip (Unzipped Files)/Images/car.jpg')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
car_detection = car_detector.detectMultiScale(image_gray, scaleFactor = 1.01, minNeighbors=4, maxSize=(60,60))
for (x, y, w, h) in car_detection:
  print(w,h)
  cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
cv2_imshow(image)
```
